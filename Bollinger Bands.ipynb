```python
import requests
import json
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
```

```python
pair = 'BTC-XVG'
duration = 10
avg_name = 'AVG_' + str(duration)
sd_name = 'SD_' + str(duration)
sd_range = 1
url = "https://bittrex.com/Api/v2.0/pub/market/GetTicks?marketName=" + pair +"&tickInterval=hour"
start = dt.datetime(2017,12,1)
end = dt.datetime(2018,1,19)
resample_key = '8H'
```

```python
r = requests.get(url)
```

```python
coin = pd.DataFrame(r.json()['result'])
```

```python
# Only need closing data , so dropping Open,Low,High,BV
coin.drop(['O','L','H','BV'],inplace=True,axis=1)
```

```python
coin.head()
```

```python
# Convert Time column frfilteredData Object to Timestamp
coin['T']=pd.to_datetime(coin['T'])
```

```python
# Set index as T
coin.set_index('T',inplace=True)
coin.head()
```

```python
resampled = coin.resample(resample_key).last()
resampled.head(10)
```

```python
filteredData = resampled[resampled.index>start].copy()
```

```python
filteredData.head()
```

```python
filteredData[avg_name] = filteredData['C'].rolling(duration).mean()
filteredData[sd_name] = filteredData['C'].rolling(duration).std()
filteredData['upper_sd'] = filteredData[avg_name] + sd_range*filteredData[sd_name]
filteredData['lower_sd'] = filteredData[avg_name] - sd_range*filteredData[sd_name]
```

```python
plt.figure(figsize=(16,10))
plt.plot(filteredData.index,filteredData['C'])
plt.plot(filteredData.index,filteredData[avg_name])
plt.plot(filteredData.index,filteredData['upper_sd'])
plt.plot(filteredData.index,filteredData['lower_sd'])
plt.legend(['Current Price','Avg','upper sd','lower sd'])
plt.show()
```

```python
filteredData.head(20)
```

```python
filteredData['BUY'] = np.where((filteredData['C'] > filteredData['lower_sd']) & (filteredData['C'].shift(1) < filteredData['lower_sd'].shift(1)),filteredData['C'],0)
```

```python
filteredData['SELL'] = np.where((filteredData['C'] > filteredData['upper_sd']) & (filteredData['C'].shift(1) < filteredData['upper_sd'].shift(1)),filteredData['C'],0)
```

```python
# buysellDataframe = filteredData.copy()
buysellDataframe = filteredData[(filteredData['BUY'] > 0) | (filteredData['SELL'] > 0)]
buysellDataframe.head(20)
```

```python
plt.figure(figsize=(16,10))
plt.plot(buysellDataframe.index,buysellDataframe['C'])
plt.plot(buysellDataframe.index,buysellDataframe[avg_name])
plt.plot(buysellDataframe.index,buysellDataframe['upper_sd'],'r')
plt.plot(buysellDataframe.index,buysellDataframe['lower_sd'],'b')
plt.legend(['Current Price','Avg','upper sd','lower sd'])
plt.show()
```

```python
# Backtesting...
BTC = 1
stocks = 0.0
previousBuyPrice = 0.0
previousSellPrice = 0.0
print('Started with BTC :' + str(BTC))
for i,r in buysellDataframe.iterrows():
    
    if (r['BUY'] != 0.0) & (stocks == 0.0):
        stocks = BTC/r['BUY']
    elif (r['SELL'] != 0.0) & (stocks != 0.0):
        BTC = stocks*r['SELL']
        stocks = 0
    

print('Final ' + pair + ' bought : #' + str(stocks))
print('Made BTC # :' + str(BTC))
```
