```python
import requests
import json
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
```

```python
pair = 'BTC-XVG'
url = "https://bittrex.com/Api/v2.0/pub/market/GetTicks?marketName=" + pair +"&tickInterval=hour"
start = dt.datetime(2017,12,1)
end = dt.datetime(2018,1,19)
```

```python
r = requests.get(url)
```

```python
coin = pd.DataFrame(r.json()['result'])
```

```python
# Only need closing data , so dropping Open,Low,High,BV
coin.drop(['O','L','H','BV'],inplace=True,axis=1)
```

```python
coin.head()
```

```python
# Convert Time column frfilteredData Object to Timestamp
coin['T']=pd.to_datetime(coin['T'])
```

```python
# Set index as T
coin.set_index('T')
```

```python
filteredData = coin[coin['T']>start].copy()
```

```python
filteredData.head()
```

```python
filteredData['SMA9'] = filteredData['C'].rolling(9).mean()
filteredData['SMA21'] = filteredData['C'].rolling(21).mean()
```

```python
plt.figure(figsize=(16,10))
plt.plot(filteredData['T'],filteredData['C'])
plt.plot(filteredData['T'],filteredData['SMA9'])
plt.plot(filteredData['T'],filteredData['SMA21'])
plt.legend(['Current Price','SMA9','SMA21'])
plt.show()
```

```python
filteredData.head()
```

```python
filteredData['BUY'] = np.where((filteredData['SMA9'] > filteredData['SMA21']) & (filteredData['SMA9'].shift(1) < filteredData['SMA21'].shift(1)),filteredData['C'],0)
```

```python
filteredData['SELL'] = np.where((filteredData['SMA9'] < filteredData['SMA21']) & (filteredData['SMA9'].shift(1) > filteredData['SMA21'].shift(1)),filteredData['C'],0)
```

```python
buysellDataframe = filteredData[(filteredData['BUY'] > 0) | (filteredData['SELL'] > 0)]
buysellDataframe.head(100)
```

```python
plt.figure(figsize=(16,10))
plt.plot(buysellDataframe['T'],buysellDataframe['C'])
plt.plot(buysellDataframe['T'],buysellDataframe['SMA9'])
plt.plot(buysellDataframe['T'],buysellDataframe['SMA21'])
plt.legend(['SMA9','SMA21'])
plt.show()
```

```python
# Backtesting...
BTC = 1
stocks = 0
print('Started with BTC :' + str(BTC))
for i,r in buysellDataframe.iterrows():
    
    if r['SELL'] == 0:
        stocks = BTC/r['BUY']
        print('Bought ' + pair + ' # ' + str(stocks) + ' on ' + str(r['T']))
    else:
        BTC = stocks*r['SELL']
        print('Made ' + str(BTC) + ' BTC on ' + str(r['T']))
    

print('Final ' + pair + ' bought : #' + str(stocks))
print('Made BTC # :' + str(BTC))
```
